```python
import re
import logging
import pandas as pd

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup, ResultSet
from functools import partial
from datetime import datetime
import time

logging.basicConfig(level=logging.INFO)
```

```python
pool = [
   'wp',    # Win Place
   'wpq',   # Q PQ
   'fct',   # 二重彩
   'tce',   # 三重彩
   'tri',   # Trio
   'ff',    # 四連環
   'qtt',   # 四重彩
   'dbl',   # 孖寶
   'jck',   # 騎師王
   'tnc',   # 練馬師王
]

URL = 'https://bet.hkjc.com/ch/racing/wpq'

```

```python
driver_options = Options()
# driver_options.add_argument("--headless=new")
driver_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36")
driver = webdriver.Chrome(options=driver_options)
driver.get(URL)

while True:
   jc_data = BeautifulSoup(driver.page_source, 'html.parser')
   qp_odd = jc_data.select('div.qin-odds-table')[0].select('div[id^="qb_QIN"]')
   data = [
      dict(
         combination=item.get('id', ''),
         odd=item.span.text if qp_odd[0].span else ''
      ) for item in qp_odd
   ]
   print(pd.DataFrame(data))
   driver.refresh()
   time.sleep(1)
   
```
